# **Data Cleaning and Wrangling**


```python
# ===============================================================
# 5. DATA CLEANING
# ===============================================================

# Strip column names
df1.columns = df1.columns.str.strip().str.lower().str.replace(" ", "_")

# Convert % columns
percent_cols = [
    "renewable_energy_share_(%)",
    "fossil_fuel_dependency_(%)",
    "industrial_energy_use_(%)",
    "household_energy_use_(%)"
]

for col in percent_cols:
    df1[col] = df1[col].astype(float)
```

```python
df1.head()
```

```python
# ------------------------------------------------------------------------
# Check duplicates -------------------------------------------------------
# ------------------------------------------------------------------------

print("Duplicates:", df.duplicated().sum())
```

`Adding a column`

```python
#-----------------------------------------------------------------------------
# Energy quality score
# ----------------------------------------------------------------------------

df1["energy_quality_score"] = (
    df1["renewable_energy_share_(%)"] -
    df1["fossil_fuel_dependency_(%)"])
```

# ðŸ“Š **EDA â€” EXPLORATORY DATA ANALYSIS**

**Total Energy Consumption by Country - Top 10 (latest year)**

```python
latest_year = df1['year'].max()
top_countries = (df1[df1['year'] == latest_year]
                 .groupby('country')['total_energy_consumption_(twh)']
                 .sum()
                 .sort_values(ascending=False)
                 .head(10))
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.xlabel("Total Consumption (TWh)")
plt.title("Top 10 Countries by Energy Consumption ({})".format(latest_year))
plt.tight_layout()
plt.show()
```

```python
df_yearly = df1.groupby('year', as_index=False)['total_energy_consumption_(twh)'].sum()
```

**Total Energy consumed over years**

```python
plt.figure(figsize=(12, 6))
plt.plot(df_yearly['year'], df_yearly['total_energy_consumption_(twh)'], marker='o')
plt.xlabel('Year')
plt.ylabel('Total Global Energy Consumption (TWh)')
plt.title('Total Global Energy Consumed per Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**SCATTER RELATIONSHIPS**

```python
plt.figure(figsize=(20,7))
sns.scatterplot(
    data=df1,
    x="total_energy_consumption_(twh)",
    y="carbon_emissions_(million_tons)",
    hue="renewable_energy_share_(%)",
    palette="viridis"
)
plt.title("Energy Consumption vs Emissions")
plt.show()
```

**Renewable Energy Share Distribution by Country**

```python
import seaborn as sns
plt.figure(figsize=(12,6))
sns.boxplot(data=df1, x='country', y='renewable_energy_share_(%)',palette="pastel")
plt.xticks(rotation=90)
plt.title('Renewable Energy Share Distribution by Country')
plt.show()
```

**Fossil Fuel Dependency Trend**

```python
yearly = df1.groupby('year', as_index=False)['fossil_fuel_dependency_(%)'].sum()
plt.figure(figsize=(10,5))
plt.plot(yearly.index, yearly["fossil_fuel_dependency_(%)"], color="red")
plt.title("Fossil Fuel Dependency Trend")
plt.xlabel("Year")
plt.ylabel("Dependency (%)")
plt.grid(True)
plt.show()
```

**Fossil Fuel Use Share by Country (Top 6)**

```python
# Top 6 countries for clear pie chart
top_countries = (
    df1.groupby('country')['fossil_fuel_dependency_(%)']
    .mean()
    .sort_values(ascending=False)
    .head(6)
)

plt.figure(figsize=(6,6))
plt.pie(top_countries.values, labels=top_countries.index, autopct='%1.1f%%')
plt.title("Fossil Fuel Use Share by Country (Top 6)")
plt.show()
```

**Carbon Emissions vs Fossil Fuel Dependency**

```python
plt.figure(figsize=(20,7))
sns.scatterplot(data=df1, x='fossil_fuel_dependency_(%)', y='carbon_emissions_(million_tons)', hue='country', alpha=0.7)
plt.title('Carbon Emissions vs Fossil Fuel Dependency')
plt.show()
```

**Top 10 Countries by per_capita_energy_use_(kwh)**

```python
latest_year = df1['year'].max()
top_countries = (df1[df1['year'] == latest_year]
                 .groupby('country')['per_capita_energy_use_(kwh)']
                 .sum()
                 .sort_values(ascending=False)
                 .head(10))
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.values, y=top_countries.index,palette="deep")
plt.xlabel("per_capita_energy_use_(kwh)")
plt.title("Top 10 Countries by per_capita_energy_use_(kwh) ({})".format(latest_year))
plt.tight_layout()
plt.show()
```

**Global Energy Mix (Donut Chart)**

```python
values = [
        df1['renewable_energy_share_(%)'].mean(),
        df1['fossil_fuel_dependency_(%)'].mean()
    ]
labels = ['Renewable', 'Fossil']

plt.figure(figsize=(5,5))
wedges, texts, autotexts = plt.pie(values, labels=labels, autopct='%1.1f%%')

# Add donut circle
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Global Energy Mix (Donut Chart)")
plt.show()
```

**Renewable Energy Share Over Years**

```python
yearly = df1.groupby('year')['renewable_energy_share_(%)'].mean()

plt.figure(figsize=(8,4))
plt.plot(yearly.index, yearly.values, marker='o')
plt.title("Renewable Energy Share Over Years")
plt.xlabel("Year")
plt.ylabel("Renewable Share (%)")
plt.grid(True)
plt.show()
```

**Top 10 Countries by Industrial Energy Use (%)**

```python
df1['industrial_energy_use_(%)'] = pd.to_numeric(df1['industrial_energy_use_(%)'], errors='coerce')
data = df1.groupby('country')['industrial_energy_use_(%)'].mean().sort_values().tail(10)
plt.figure(figsize=(10,5))
plt.barh(data.index, data.values, color='skyblue')
plt.title("Top 10 Countries by Industrial Energy Use (%)")
plt.xlabel("Industrial Energy Use (%)")
plt.show()
```

**Energy Price Index Over Time**

```python
col = 'energy_price_index_(usd/kwh)'
df1[col] = pd.to_numeric(df1[col], errors='coerce')
yearly = df1.groupby('year')['energy_price_index_(usd/kwh)'].mean()
plt.figure(figsize=(8,4))
plt.plot(yearly.index, yearly.values, marker='o')
plt.title("Energy Price Index Over Time")
plt.xlabel("Year")
plt.ylabel("Price (USD/kWh)")
plt.grid(True)
plt.show()
```

**Top 10 Countries by Energy Quality Score**

```python
df1['energy_quality_score'] = pd.to_numeric(df1['energy_quality_score'], errors='coerce')
data = df1.groupby('country')['energy_quality_score'].mean().sort_values().tail(10)
plt.figure(figsize=(10,5))
plt.barh(data.index, data.values, color='darkseagreen')
plt.title("Top 10 Countries by Energy Quality Score")
plt.xlabel("Quality Score")
plt.show()
```
